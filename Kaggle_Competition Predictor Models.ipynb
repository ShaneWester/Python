```python
#Question 2
#Light gradient Boosting Machine (LGBM Regressor) is a model specifically designed for speed and performance and is made for large, high dimensional datasets like the one we have processed in Task 2. LGBM builds a range of decision trees and each tree is trained to correct the errors of the previous decision tree. We will use this model due to its computational efficiency and ability to handle categorical variables and large datasets

#The Stacking Regressor combines multiple base models to enhance predictions. The Stacking Regressor uses multiple models that will be trained on the processed dataset which will then combine their predictions. This will help reduce individual model bias and extract the strengths of each model used and hopefully increase overall accuracy

#Linear regression will be used as it is one of the simplest models to use.  It assumes a linear relationship between the input variables and the target variable, making it easy to understand and implement.It is very computationally efficient and is easy to interpret. It may not be the strongest model, but it will serve as a good benchmark to evaluate the effectiveness of other models
```

```python
#Question 3
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import StackingRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV
import numpy as np
import warnings

# Suppress specific LightGBM warnings globally
warnings.filterwarnings("ignore", message="No further splits with positive gain")

# Prepare data
X_train = train_data_selected.drop(columns=['price'])  # Features
y_train = train_data_selected['price']                 # Target variable

# 1. Linear Regression
lr = LinearRegression()
lr_scores = cross_val_score(lr, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
lr_rmse = np.sqrt(-lr_scores)
print("Linear Regression - Mean RMSE:", lr_rmse.mean(), "Std Dev RMSE:", lr_rmse.std())

# 2. LGBM Regressor with Hyperparameter Tuning
# Define LGBM with verbose=-1 to suppress warnings
lgbm = LGBMRegressor(random_state=42, verbose=-1)

# Define parameter grid for LGBM
param_grid_lgbm = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 6, 9]
}

# Perform GridSearchCV for LGBM
lgbm_grid_search = GridSearchCV(estimator=lgbm, param_grid=param_grid_lgbm, 
                                cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
lgbm_grid_search.fit(X_train, y_train)

# Best model and parameters for LGBM
lgbm_best = lgbm_grid_search.best_estimator_
print("LGBM Best Parameters:", lgbm_grid_search.best_params_)

# Evaluate LGBM using cross-validation
lgbm_scores = cross_val_score(lgbm_best, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
lgbm_rmse = np.sqrt(-lgbm_scores)
print("LGBM Regressor - Mean RMSE:", lgbm_rmse.mean(), "Std Dev RMSE:", lgbm_rmse.std())

# 3. Stacking Regressor
# Define base models for stacking
cat_model = CatBoostRegressor(iterations=1000, learning_rate=0.05, depth=6, random_seed=42, silent=True)
lgb_model = LGBMRegressor(n_estimators=1000, learning_rate=0.05, max_depth=6, random_state=42, verbose=-1)  # Suppressed

# Define Stacking Regressor with Ridge as the meta-model
stacking_regressor = StackingRegressor(
    estimators=[
        ('cat', cat_model),
        ('lgb', lgb_model)
    ],
    final_estimator=Ridge(alpha=1.0)
)

# Evaluate Stacking Regressor using cross-validation
stacking_scores = cross_val_score(stacking_regressor, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
stacking_rmse = np.sqrt(-stacking_scores)
print("Stacking Regressor - Mean RMSE:", stacking_rmse.mean(), "Std Dev RMSE:", stacking_rmse.std())
```

```python
#For Linear Regression, there were no hyperparameters to tune, and it resulted in a high RMSE of 9930.03. This indicates that the model did not fit the data well, likely because it is too simple for the datasets complexity.

#For the LGBM Regressor, key hyperparameters such as learning_rate=0.1, max_depth=9, and n_estimators=500 were tuned. This brought the RMSE down to 6170.58, which is a significant improvement over Linear Regression, suggesting that LGBM handled the complexity better.

#Finally, the Stacking Regressor, which combines CatBoost and LGBM with Ridge as the meta-model, achieved the best performance with an RMSE of 6093.47. This demonstrates that stacking multiple models can improve accuracy by capturing different patterns in the data.

#In summary, the Stacking Regressor had the lowest RMSE, making it the most accurate model on this dataset.
```

```python
#Question 4 : LGBM-REGRESSOR

import pandas as pd
import lightgbm as lgb

# Load the train and test datasets
train_data = pd.read_csv('train_processed_imputed.csv')
test_data = pd.read_csv('test_processed_imputed.csv')

# Separate 'vin' column from the test data to use as ID later
vin_column = test_data['vin']  # Save vin for submission

# Drop 'vin' from both train and test data to ensure it's not part of the model features
train_data = train_data.drop(columns=['vin'])
test_data = test_data.drop(columns=['vin'])

# Columns that need conversion or cleaning
non_numeric_columns = ['city', 'listed_date', 'maximum_seating', 'power', 'torque']

# Function to extract numeric values from columns like 'power' and 'torque'
def extract_numeric(column):
    return pd.to_numeric(column.str.extract('(\d+\.?\d*)')[0], errors='coerce')

# Apply extraction function to columns with embedded numbers
for col in ['power', 'torque']:
    train_data[col] = extract_numeric(train_data[col])
    test_data[col] = extract_numeric(test_data[col])

# One-hot encode categorical columns (excluding 'vin')
train_data = pd.get_dummies(train_data, columns=['city', 'listed_date', 'maximum_seating'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['city', 'listed_date', 'maximum_seating'], drop_first=True)

# Align columns in test set with training set
test_data = test_data.reindex(columns=train_data.columns, fill_value=0)

# Define features and target variable
X_train = train_data.drop(columns=['price'])  # Feature columns, excluding target
y_train = train_data['price']                 # Target variable
X_test = test_data[X_train.columns]           # Ensure test data has matching columns

# LightGBM model definition
lgb_model = lgb.LGBMRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

# Train the model
lgb_model.fit(X_train, y_train)

# Make predictions on the test set
lgb_predictions = lgb_model.predict(X_test)

# Prepare the submission DataFrame, using the saved 'vin' column as ID
submission = pd.DataFrame({
    'vin': vin_column,      # Use the saved 'vin' as the ID column
    'PredictedPrice': lgb_predictions
})

# Save the submission as 'submission_lightgbm.csv'
submission.to_csv('submission_lightgbm.csv', index=False)

# Display the first few rows of the submission file to verify
submission.head()
```

```python
#CATBOOSTREGRESSOR

from catboost import CatBoostRegressor

# Load the train and test datasets
train_data = pd.read_csv('train_processed_imputed.csv')
test_data = pd.read_csv('test_processed_imputed.csv')

# Define features and target variable
X_train = train_data.drop(columns=['price'])  # Use only feature columns
y_train = train_data['price']                 # Target variable

# Prepare test features, ensuring they match the training features
X_test = test_data[X_train.columns]           # Match columns to training data

# Identify categorical columns by checking data types
# Assuming any column with object type is categorical
categorical_features = X_train.select_dtypes(include=['object']).columns.tolist()

# Convert categorical feature names to indices (required by CatBoost)
categorical_feature_indices = [X_train.columns.get_loc(col) for col in categorical_features]

# Define and train the CatBoost model, specifying categorical features
cat_model = CatBoostRegressor(
    iterations=1000, learning_rate=0.05, depth=6, random_seed=42, silent=True,
    cat_features=categorical_feature_indices  # Specify categorical features here
)
cat_model.fit(X_train, y_train)

# Make predictions on the test set
cat_predictions = cat_model.predict(X_test)


# Prepare the submission DataFrame
submission = pd.DataFrame({
    'vin': test_data['vin'],      # Use the 'vin' column as ID
    'PredictedPrice': cat_predictions
})

# Save the submission to a CSV file
submission.to_csv('submission.csv', index=False)

# Display the first few rows to verify
submission.head()
```

```python
#STACKING-REGRESSOR

import pandas as pd
from catboost import CatBoostRegressor
import lightgbm as lgb
from sklearn.linear_model import Ridge
from sklearn.ensemble import StackingRegressor

# Load the train and test datasets
train_data = pd.read_csv('train_processed_imputed.csv')
test_data = pd.read_csv('test_processed_imputed.csv')

# Suppress specific LightGBM warnings globally
warnings.filterwarnings("ignore", message="No further splits with positive gain")

# Save and remove 'vin' from test data for submission
vin_column = test_data['vin']
train_data = train_data.drop(columns=['vin'])
test_data = test_data.drop(columns=['vin'])

# Columns with non-numeric data that need to be converted
non_numeric_columns = ['city', 'listed_date', 'maximum_seating', 'power', 'torque']

# Function to extract numeric values from columns like 'power' and 'torque'
def extract_numeric(column):
    return pd.to_numeric(column.str.extract('(\d+\.?\d*)')[0], errors='coerce')

# Apply extraction function to columns with embedded numbers
for col in ['power', 'torque']:
    train_data[col] = extract_numeric(train_data[col])
    test_data[col] = extract_numeric(test_data[col])

# One-hot encode categorical columns
train_data = pd.get_dummies(train_data, columns=['city', 'listed_date', 'maximum_seating'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['city', 'listed_date', 'maximum_seating'], drop_first=True)

# Align columns in test set with training set
test_data = test_data.reindex(columns=train_data.columns, fill_value=0)

# Define features and target variable
X_train = train_data.drop(columns=['price'])  # Feature columns, excluding target
y_train = train_data['price']                 # Target variable
X_test = test_data[X_train.columns]           # Ensure test data has matching columns

# Define the base models
cat_model = CatBoostRegressor(iterations=1000, learning_rate=0.05, depth=6, random_seed=42, silent=True)
lgb_model = lgb.LGBMRegressor(n_estimators=1000, learning_rate=0.05, max_depth=6, random_state=42)

# Define the stacking model with Ridge as the final estimator
stacking_model = StackingRegressor(
    estimators=[
        ('cat', cat_model),
        ('lgb', lgb_model)
    ],
    final_estimator=Ridge(alpha=1.0)
)

# Train the stacking model
stacking_model.fit(X_train, y_train)

# Make predictions on the test set using the stacking model
stacking_predictions = stacking_model.predict(X_test)

# Prepare submission DataFrame with stacking predictions
submission = pd.DataFrame({
    'vin': vin_column,  # Use the saved 'vin' as the ID column
    'PredictedPrice': stacking_predictions
})

# Save the submission file
submission.to_csv('submission_stacking.csv', index=False)
submission.head()
```
