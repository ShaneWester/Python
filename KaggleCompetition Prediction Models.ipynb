```python
import pandas as pd

# datasets
train_processed = pd.read_csv('train_processed.csv')
test_processed = pd.read_csv('test_processed.csv')

# function to extract numeric values from columns with text
def extract_numeric(column):
    return pd.to_numeric(column.str.extract(r'(\d+\.?\d*)')[0], errors='coerce')

# Convert specific columns to numeric
for col in ['back_legroom', 'front_legroom', 'power', 'torque']:
    if col in train_processed.columns and col in test_processed.columns:
        train_processed[col] = extract_numeric(train_processed[col].astype(str))
        test_processed[col] = extract_numeric(test_processed[col].astype(str))
```

`In this part, we start by loading our training and test datasets from CSV files. We want to make sure that the features we use for our model are all in a numerical format, especially since some of them might contain both numbers and text. To do this, we create a function called extract_numeric. This function uses a regular expression to pull out the numbers from the columns we’re interested in, like back_legroom, front_legroom, power, and torque. After defining this function, we apply it to those specific columns in both the training and test datasets.`

```python
# 1. Split torque and power into value and RPM 
train_processed['torque_value'] = train_processed['torque']
train_processed['power_value'] = train_processed['power']
test_processed['torque_value'] = test_processed['torque']
test_processed['power_value'] = test_processed['power']

# 2. Calculate Car Age based on listed_date and year
if 'listed_date' in train_processed.columns and 'year' in train_processed.columns:
    train_processed['car_age'] = pd.to_datetime(train_processed['listed_date']).dt.year - train_processed['year']
    test_processed['car_age'] = pd.to_datetime(test_processed['listed_date']).dt.year - test_processed['year']

# 3. Fuel Economy Difference
train_processed['fuel_economy_diff'] = train_processed['city_fuel_economy'] - train_processed['highway_fuel_economy']
test_processed['fuel_economy_diff'] = test_processed['city_fuel_economy'] - test_processed['highway_fuel_economy']

# 4. Engine Size Grouping
def categorize_engine_size(displacement):
    if displacement < 2.0:
        return 'small'
    elif 2.0 <= displacement < 3.5:
        return 'medium'
    else:
        return 'large'

train_processed['engine_size_group'] = train_processed['engine_displacement'].apply(categorize_engine_size)
test_processed['engine_size_group'] = test_processed['engine_displacement'].apply(categorize_engine_size)

# One-hot encode 'engine_size_group'
train_processed = pd.concat([train_processed, pd.get_dummies(train_processed['engine_size_group'], prefix='engine_size')], axis=1)
test_processed = pd.concat([test_processed, pd.get_dummies(test_processed['engine_size_group'], prefix='engine_size')], axis=1)
```

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately. For instance, if appropriate, you could all map colours into {red, yellow, green, blue, purple, other} and then encode.  
(2 marks)

```python
# Combine 'interior_color' columns into a single 'interior_color' column
def combine_color_columns(df, color_columns_prefix='interior_color_'):
    color_columns = [col for col in df.columns if col.startswith(color_columns_prefix)]
    def get_color(row):
        for col in color_columns:
            if row[col] == 1:
                return col.replace(color_columns_prefix, '')
        return 'Other'
    df['interior_color'] = df.apply(get_color, axis=1)
    df.drop(columns=color_columns, inplace=True)

combine_color_columns(train_processed)
combine_color_columns(test_processed)

# Limit 'city' and 'interior_color' to top 5 categories, replacing others with 'Other'
def encode_top_categories(df, column, top_n=5):
    top_categories = df[column].value_counts().nlargest(top_n).index
    df[column] = df[column].apply(lambda x: x if x in top_categories else 'Other')

for col in ['city', 'interior_color']:
    encode_top_categories(train_processed, col, top_n=5)
    encode_top_categories(test_processed, col, top_n=5)

# Apply one-hot encoding to 'city' and 'interior_color'
train_processed = pd.get_dummies(train_processed, columns=['city', 'interior_color'], drop_first=True)
test_processed = pd.get_dummies(test_processed, columns=['city', 'interior_color'], drop_first=True)

# Align test_processed to match train_processed columns
test_processed = test_processed.reindex(columns=train_processed.columns, fill_value=0)
```

`Combining Interior Color Columns: 
In this section, we define a function called combine_color_columns, which merges multiple interior color columns into a single column named interior_color. This function checks each row to see if any of the specified color columns have a value of 1. If so, it replaces the combined color values with "Other" for those that do not fit the specified categories. This step simplifies the dataset and prepares it for encoding.`

`Limiting Categories: 
Next, we define the encode_top_categories function, which limits categorical features, specifically city and interior_color, to their top five categories. Any other categories that exceed this limit are replaced with the label "Other." This is crucial because it reduces the dimensionality of the categorical variables and improves the efficiency of the model. By focusing on the most common categories, we ensure that the model is not overwhelmed by rare values, which can introduce noise.`

`Applying One-Hot Encoding: 
After limiting the categories, we apply one-hot encoding to the city and interior_color columns using pd.get_dummies(). This technique transforms categorical variables into a format that can be provided to machine learning algorithms to improve prediction accuracy. One-hot encoding creates new binary columns for each category, indicating whether each observation belongs to that category (1) or not (0).`

`Aligning Test Data: 
Finally, we align the test dataset with the training dataset by ensuring that both datasets have the same columns after one-hot encoding.
```python
from sklearn.preprocessing import StandardScaler

# Define numerical columns for scaling (excluding target variable 'price' in test data)
numerical_columns = train_processed.select_dtypes(include=['float64', 'int64']).columns
if 'price' in numerical_columns:
    numerical_columns = numerical_columns.drop('price')  # Exclude target variable if it's in train data

# 1. Scale Numerical Features to Standardise
scaler = StandardScaler()
train_processed[numerical_columns] = scaler.fit_transform(train_processed[numerical_columns])
test_processed[numerical_columns] = scaler.transform(test_processed[numerical_columns])

# 2. Cap Outliers for Specified Columns
def cap_outliers(df, column, factor=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)

# Outlier capping to specific columns in the training data
cap_outliers(train_processed, 'mileage')
cap_outliers(train_processed, 'horsepower')
cap_outliers(train_processed, 'price')  # Only cap 'price' in the training set

# Saving final processed datasets
train_processed.to_excel('train_processed_final.xlsx', index=False)
test_processed.to_excel('test_processed_final.xlsx', index=False)
```
